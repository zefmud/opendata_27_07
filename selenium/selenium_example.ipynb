```python
from pyquery import PyQuery as pq
```

```python
EX_MPS_PAGE = "http://w1.c1.rada.gov.ua/pls/site2/fetch_mps?skl_id=9&pid_id=-3"
```

```python
page = pq(EX_MPS_PAGE)
```

```python
mp_links = page("p.title a")
```

```python
names = [pq(l).text() for l in mp_links]
links = [pq(l).attr("href") for l in mp_links]
```

```python
print(names)
```

```python
print(links)
```
